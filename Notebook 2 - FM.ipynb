## Código Resuelto

```python
#Importamos librerías
```

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
from scipy.io.wavfile import write
```

```python
#Parámetros de la señal
```

```python
mps= 1000 #muestras por segundo
frec_portadora = 100  #hz
frec_moduladora = 1 #hz
duracion = 2.5 #segundos
k=15.0  #indice de modulacion
```

```python
#Calculamos la señal sinusoidal de la portadora FM
```

```python
muestras= frec_moduladora/mps
t=np.arange(0, 1, muestras)
```

```python
#Calculamos la señal sinusoidal de la modulante y la señal modulada FM
```

```python
moduladora = np.sin(2 * np.pi * frec_moduladora* t)
```

```python
señal_modulada_FM = np.sin(2*np.pi*(frec_portadora+k*moduladora)*t)
```

```python
plt.plot(t,señal_modulada_FM)
```
