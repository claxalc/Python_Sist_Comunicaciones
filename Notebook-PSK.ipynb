```python
#Importar librerías
```

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
#Parámetros de la señal
```

```python
señal_digital=[0,1,1,0,0,1,0,1,1,1,0,1]   #Señal digital 
```

```python
número_muestras=200     #número de muestras
```

```python
#Toma de Muestras
```

```python
señal_muestreada=[]
```

```python
for i in range(1,12):    #Porque tenemos 12 dígitos en la señal digital
    u=np.ones(200)  #Vector lleno de unos que son los pulsos para tomar muestras
    m=u*señal_digital[i]    #muestras
    señal_muestreada=np.concatenate((señal_muestreada,m))
```

```python
plt.subplot(4,1,1)
plt.plot(señal_muestreada)    #Esta es la señal modulante (binaria)
```

```python
#señal sinusoidal de la portadora:
```

```python
dimension=len(señal_muestreada)   #es basicamente la longitud del vector señal_muestreada
```

```python
t=np.linspace(0,5,dimension)   #periodo de la portadora
```

```python
frec=7  #frecuencia de la portadora
```

```python
portadora_sen=np.sin(2*np.pi*frec*t)   #Señal portadora seno
```

```python
portadora_cos=np.cos(2*np.pi*frec*t)   #Señal portadora coseno
```

```python
plt.subplot(4,1,2)
plt.plot(t,portadora_sen)
plt.subplot(4,1,3)
plt.plot(t,portadora_cos)
```

```python
#Señal modulada:
```

```python
señal_modulada =[]
for i in range(0,dimension):
    if señal_muestreada[i]==0:
        posicion_cero=np.array([portadora_sen[i]])
        señal_modulada=np.concatenate((señal_modulada,posicion_cero))
    else:
        posicion_uno=np.array([portadora_cos[i]])
        señal_modulada=np.concatenate((señal_modulada,posicion_uno))
```

```python
plt.subplot(4,1,4)
plt.plot(t,señal_modulada)
```

### ¿En qué consiste la modulación PSK?

La modulación por desplazamiento de fase o PSK (Phase Shift Keying) es una forma de modulación angular que consiste en hacer variar la fase de la
portadora entre un número de valores discretos. Dependiendo del número de posibles fases a tomar,recibe diferentes denominaciones. Dado que lo más
común es codificar un número entero de bits por cada símbolo, el número de fases a tomar es una potencia de dos.
