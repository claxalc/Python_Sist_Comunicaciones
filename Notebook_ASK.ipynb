## Código Resuelto

```python
#Importar librerías
```

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
#Parámetros de la señal
```

```python
señal_digital=[0,1,1,0,0,1,0,1,1,1,0,1]   #Señal digital 12 digitos
```

```python
numero_muestras=200   #número de muestras
```

```python
#Toma de Muestras
```

```python
señal_muestreada=[]
```

```python
for i in range(1,12):    #Porque tenemos 12 dígitos en la señal digital
    u=np.ones(numero_muestras)  #Vector lleno de unos que son los pulsos para tomar muestras  (Reloj)
    m=u*señal_digital[i]    #muestras
    señal_muestreada=np.concatenate((señal_muestreada,m))
```

```python
plt.subplot(3,1,1)
plt.plot(señal_muestreada)    #Esta es la señal modulante (binaria)
```

```python
#señal sinusoidal de la portadora:
```

```python
dimension=len(señal_muestreada)   #es basicamente la longitud del vector señal_muestreada
```

```python
t=np.linspace(0,5,dimension)   #periodo de la portadora
```

```python
frec_p=5   #frecuencia de la portadora en hertz
```

```python
señal_portadora=np.cos(2*np.pi*frec_p*t)
```

```python
plt.subplot(3,1,2)
plt.plot(t,señal_portadora)
```

```python
#Señal modulada:
```

```python
señal_modulada=señal_muestreada*señal_portadora
```

```python
plt.subplot(3,1,3)
plt.plot(t,señal_modulada)
```

```python
plt.subplot(3,1,1)
plt.plot(señal_muestreada)  
plt.subplot(3,1,2)
plt.plot(t,señal_portadora)
plt.subplot(3,1,3)
plt.plot(t,señal_modulada)
```

### ¿En qué consiste la modulación ASK?

La modulación por desplazamiento de amplitud, Amplitude-shift keying (ASK), es una forma de modulación en la cual se
representan los datos digitales como variaciones de amplitud de la onda portadora. La amplitud de una señal portadora análoga varía conforme a la corriente de bit (modulando la señal), manteniendo la frecuencia y la fase constante. El nivel de amplitud puede ser usado para representar los valores binarios 0s y 1s.

### ¿Qué operación matemática representa a la modulación ASK? ¿Esta operación es conmutativa?

La multiplicación y si es conmutativa es decir que se puede intercambiar el orden entre la modulante (señal binaria) y la portadora.

### ¿Qué sucede si el número de muestras que tomas es igual a 3? Captura y copia la gráfica resultante y describe lo que pasó

```python


#Parámetros de la señal

señal_digital=[0,1,1,0,0,1,0,1,1,1,0,1]   #Señal digital 12 digitos

numero_muestras=3   #número de muestras
```

```python
#Toma de Muestras

señal_muestreada=[]

for i in range(1,12):    #Porque tenemos 12 dígitos en la señal digital
    u=np.ones(numero_muestras)  #Vector lleno de unos que son los pulsos para tomar muestras  (Reloj)
    m=u*señal_digital[i]    #muestras
    señal_muestreada=np.concatenate((señal_muestreada,m))

plt.subplot(3,1,1)
plt.plot(señal_muestreada)    #Esta es la señal modulante (binaria)
```

```python
#señal sinusoidal de la portadora:

dimension=len(señal_muestreada)   #es basicamente la longitud del vector señal_muestreada

t=np.linspace(0,5,dimension)   #periodo de la portadora

frec_p=5   #frecuencia de la portadora en hertz


```

```python
señal_portadora=np.cos(2*np.pi*frec_p*t)


plt.subplot(3,1,2)
plt.plot(t,señal_portadora)
```

```python
señal_modulada=señal_muestreada*señal_portadora

plt.subplot(3,1,3)
plt.plot(t,señal_modulada)
```

Podemos observar que si el número de muestras en muy bajo no es posible muestrear la señal modulante de forma adecuada por tanto la modulación tampoco podrá ser efectuada.
