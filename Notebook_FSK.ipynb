## Código Resuelto

```python
#Importar librerías
```

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
#Parámetros de la señal
```

```python
señal_digital=[0,1,1,0,0,1,0,1,1,1,0,1]   #Señal digital 
```

```python
número_muestras=200     #número de muestras
```

```python
#Toma de Muestras
```

```python
señal_muestreada=[]
```

```python
for i in range(1,12):    #Porque tenemos 12 dígitos en la señal digital
    u=np.ones(200)  #Vector lleno de unos que son los pulsos para tomar muestras
    m=u*señal_digital[i]    #muestras
    señal_muestreada=np.concatenate((señal_muestreada,m))
```

```python
plt.subplot(4,1,1)
plt.plot(señal_muestreada)    #Esta es la señal modulante (binaria)
```

```python
#señal sinusoidal de la portadora:
```

```python
dimension=len(señal_muestreada)   #es basicamente la longitud del vector señal_muestreada
```

```python
t=np.linspace(0,5,dimension)   #periodo de la portadora
```

```python
frec_s=5   #frecuencia espacio
```

```python
frec_m=25  #Frecuencia marca
```

```python
portadora_s=np.cos(2*np.pi*frec_s*t)   #Señal portadora con frecunencia fs
```

```python
portadora_m=np.cos(2*np.pi*frec_m*t)   #Señal portadora con frecunencia fm
```

```python
plt.subplot(4,1,2)
plt.plot(t,portadora_s)
plt.subplot(4,1,3)
plt.plot(t,portadora_m)
```

```python
#Señal modulada:
```

```python
señal_modulada =[]
for i in range(0,dimension):
    if señal_muestreada[i]==0:
        posicion_cero=np.array([portadora_s[i]])
        señal_modulada=np.concatenate((señal_modulada,posicion_cero))
    else:
        posicion_uno=np.array([portadora_m[i]])
        señal_modulada=np.concatenate((señal_modulada,posicion_uno))
```

```python
plt.subplot(4,1,4)
plt.plot(t,señal_modulada)
```

```python
plt.subplot(4,1,1)
plt.plot(señal_muestreada)  
plt.subplot(4,1,2)
plt.plot(t,portadora_s)
plt.subplot(4,1,3)
plt.plot(t,portadora_m)
plt.subplot(4,1,4)
plt.plot(t,señal_modulada)
```

## Preguntas:

### Explica en que consiste la modulación FSK

Consiste en variar la frecuencia de la portadora de acuerdo a los datos.
Para “1” lógico le corresponde una frecuencia f1 y para un “0” lógico, emplea una frecuencia f2. 

### ¿Por qué es necesario utilizar dos frecuencias en FSK?

Precisamente para representar la señal binaria que tiene dos valores que son 0 y 1

### ¿Qué relación existe entre la frecuencia de la portadora, la frecuencia espacio y la frecuencia marca?

La frecuencia portadora tiene dos valores: la frecuencia espacio para representar al cero y la frecuencia marca para representar al uno.
